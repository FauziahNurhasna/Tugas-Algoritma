```python
#In class session
import math
print('      IS IT PRIME?    ')
print('=========================')
print()

Divisible = []

Number = int(input('Input a number:'))
print()

if Number < 0:
    print("There are no negative prime numbers.")

elif Number == 0:
    print("0 is neither prime nor composite.")

elif Number == 1:
    print("1 is NOT a PRIME number.")
    
else:
    is_prime = True
    Round = round(math.sqrt(Number)) + 1
    for i in range(2, Round):
        if Number % i == 0:
            is_prime = False
            break
    
    if is_prime:
        print('{} is a PRIME number'.format(Number))
    else:
        print('{} is NOT a PRIME number'.format(Number))
        for num in range(1, Number):
             if Number % num == 0:
                Divisible.append(num)
        print('{} is divisible by {}'.format(Number,Divisible))
```

```python
#TUGAS : Mencari bilangan prima antara 1000n dan 1000(n+1)
#NIM akhir = 027
#Maka kita akan mencari bilangan-bilangan prima antara 270000 dan 28000

import math

print('BILANGAN PRIMA ANTARA 27000 DAN 28000')
print('=====================================')
print()

Prima_27 = [] #Membuat list untuk menyetor bilangan prima

#Membuat loop untuk membatasi bilangan prima antara angka 27000 dan 28000
for i in range(27000, 28001):
    BilPrima_27 = True
    if i > 1:  # Karena bilangan prima lebih besar dari 1
        #Cek setiap angka di range (27000, 28001) untuk angka prima
        for j in range(2, int(math.sqrt(i)) + 1):
            if i % j == 0:
                BilPrima_27 = False
                break
        if BilPrima_27:
            Prima_27.append(i) #Angka yang sudah dipastikan prima disetor ke List
            
            
#Print berapa jumlah bilangan prima dan bilangan-bilangan prima itu sendiri melalui list
print("Jumlah bilangan prima antara 27000 dan 28000 adalah {}".format(len(Prima_27))) 
print()
print("Bilangan Prima: {}.".format(Prima_27))
            
```
