```python
#importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

#User inputs the coordinates for the first line and the second line
print("Coordinates for the first line.")
print("Input coordinate for the first point:")
a_27 = int(input("x coordinate:")) #for (a,b) 
b_27 = int(input('y coordinate:'))
print("Input coordinates for the second point:")
c_27 = int(input("x coordinate:")) #for (c.d)
d_27 = int(input("y coordinate:"))
print()

print("Coordinates for the second line.")
print("Input coordinate for the first point:")
e_27 = int(input("x coordinate:")) #for (e,f)
f_27 = int(input("y coordinate:"))
print("input coordinates for the second point")
g_27 = int(input("x coordinate:")) #for (g,h) 
h_27 = int(input("y coordinate:"))

#Using equations to define the gradients and constants
m1_27 = (d_27 - b_27) / (c_27 - a_27)
m2_27 = (h_27 - f_27) / (g_27 - e_27)
c1_27 = b_27 - m1_27 * a_27
c2_27 = f_27 - m2_27 * e_27

#inserting the previous gradients and constants value to a standard form for line equation and printing it
print()
print("The line equation for the first line is y={}x+{} ". format(m1_27,c1_27))
print("Thus, the gradient for the first line is " + str(m1_27) + ".")
print("The line equation for the second line is y={}x+{} ". format(m2_27,c2_27))
print("Thus, the gradient for the second line is " + str(m2_27) + ".")
print()

#Using if-else function to check whether the two lines are coincident,parallel,or intersect
if m1_27 == m2_27 and c1_27 == c2_27:
    print("The two lines are coincident.")
    x_27 = np.arange(-10,10,0.01)
    plt.plot(x_27, m1_27 * (x_27 - a_27) + b_27, color="green", label='y={}x+{}'. format(m1_27,c1_27))
    plt.plot(x_27, m2_27 * (x_27 - e_27) + f_27, color="orange", label='y={}x+{}'. format(m2_27,c2_27))
    plt.legend(loc='best')
    plt.show() 
    
elif m1_27 == m2_27 and c1_27 != c2_27:
    print("The two lines are parallel.")
    x_27 = np.arange(-10,10,0.01)
    plt.plot(x_27, m1_27 * (x_27 - a_27) + b_27, color='green', label='y={}x+{}'.format(m1_27,c1_27))
    plt.plot(x_27, m2_27 * (x_27 - e_27) + f_27, color='orange', label='y={}x+{}'. format(m2_27,c2_27))
    plt.legend(loc='best')
    plt.show()
    
else:
    print("The two lines intersect.")
    #defining the intersection
    TPx_27 = (c2_27 - c1_27) / (m1_27 - m2_27)
    TPy_27 = m1_27 * TPx_27 + c1_27
    
    print("Its intersection point is ({},{})". format(TPx_27,TPy_27))
    x_27 = np.arange(-10,10,0.01)
    plt.plot(x_27, m1_27 * (x_27 - a_27) + b_27, color='green', label='y={}x+{}'.format(m1_27,c1_27))
    plt.plot(x_27, m2_27 * (x_27 - e_27) + f_27, color='orange',label='y={}x+{}'. format(m2_27,c2_27))
    plt.plot(TPx_27,TPy_27, color='blue', marker='o', markersize=7)
    plt.legend(loc='best')
    plt.show()

```
