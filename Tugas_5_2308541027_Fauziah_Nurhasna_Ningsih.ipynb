```python
#Code for 2 dimentional list and Range

#Matrix addition function
def MatrixAddition_27 (Matrix1_27, Matrix2_27):
    Result_27 = []
    for i in range(len(Matrix1_27)):
        row = []
        for j in range(len(Matrix1_27[0])):
            row.append(Matrix1_27[i][j] + Matrix2_27[i][j])
        Result_27.append(row)
    return Result_27

#Identify Matrix 1
Matrix1_27 = [
    [2,3,4],
    [3,4,5],
    [4,5,6]
]

#Identify Matrix 2
Matrix2_27 = [
    [1,3,5],
    [7,9,11],
    [13,15,17]
]

#Getting result by applying Matrix addition function
Result_27 = MatrixAddition_27(Matrix1_27, Matrix2_27)

#Displaying result per row
print(f"The result of Matrix 1 + Matrix 2 is:")
for row in Result_27:
    print(row)
```
