```python
import math

print("Problem:")
print("x fulfills the equation x^2 + x - 1 = 0")
print("Given the equation:")
print("x^2024 - x^2025 = (x(1-x))^n")
print("Find the value of n.")
print()
print("Solution:")

print("By using logarithmic rules,we find that n = log x^2024-x^2025 / log x(1-x).")
print("Thus before finding n, we must first find the value of x by using the quadratic formula.")
print()
print("The value of x is:")

#Quadratic formula : (-b(+-)\sqrt(b^(2)-4ac))/(2a)
x1_27 = (-1 + math.sqrt(1**2 - 4*1*-1)) / (2*1)
x2_27 = (-1 - math.sqrt(1**2 - 4*1*-1)) / (2*1)

print('x1 = {}'. format(x1_27))
print('x2 = {}'. format(x2_27))

print()
Repeat_27 = True
while Repeat_27:
    print()
    print("Which of the x values will you use?")
    print('[1] x1')
    print('[2] x2')
    Choice_27 = int(input("Your answer: "))

    if Choice_27 == 1:
        print()
        print('x = {}'.format(x1_27))
        
        # finding the value of log x^2024 - x^2025 as the numerator
        # and log x(1-x) as the denominator
        # change its form so that the program could work with it
        # By following a few logarithmic rules,namely:
        # log ab = log a + log b
        # log a^b = b * log a
        Numer1_27 = 2024 * math.log(abs(x1_27)) + math.log(abs(1 - x1_27))
        Denom1_27 = math.log(abs(x1_27)) + math.log(abs(1 - x1_27))
        # .abs() ensures the result is always positive because the logarithm of a negative number is undefined
        n1_27 = Numer1_27 / Denom1_27
        print("n = {:.2f}".format(n1_27))
        print()
        Answer_27 = input('Try the other x value? [Y/N]: ')
        if Answer_27.lower() == 'n':
            Repeat_27 = False
            
    elif Choice_27 == 2:
        print()
        print('x = {}'.format(x2_27))
        
        # Same rules as before, just with different value
        Numer2_27 = 2024 * math.log(abs(x2_27)) + math.log(abs(1 - x2_27))
        Denom2_27 = math.log(abs(x2_27)) + math.log(abs(1 - x2_27))
        # .abs() ensures the result is always positive because the logarithm of a negative number is undefined
        n2_27 = Numer2_27 / Denom2_27
        print("n = {:.2f}".format(n2_27))
        print()
        Answer_27 = input('Try the other x value? [Y/N]: ')
        if Answer_27.lower() == 'n':
            Repeat_27 = False
            
    else:
        print()
        print('*sigh* No. Try again.')

```
