```python
string = 'BMI & Daily Calorie Needs Calculator'
string2 = string.center(30, '*')
print(string2)
print("What would you like to know today?")
print("[1]BMI")
print("[2]Daily Calorie Needs")
options_27 = input('Your answer: ')


while options_27 > '2' or options_27 < '1':
    print("Input invalid. Please try again.")
    options_27 = input('Your answer: ')


if options_27 == '1':
    print("To know your BMI, please input the following information: ")
    Height_27 = float(input('Height(m): '))
    Weight_27 = float(input('Weight(kg): '))
    userbmi_27 = float(Weight_27/(Height_27*Height_27))
    print('Your BMI is: ')
    print(userbmi_27)
    if userbmi_27 < 18.5:
        print("You are underweight.")
    if 18.5 <= userbmi_27 <= 24.9:
        print("You are of healthy weight.")
    if 25 <= userbmi_27 <= 29.9:
        print("You are overweight.")
    if userbmi_27 >= 30:
        print("You are obese.")


if options_27 == '2':
    print("To know your daily calorie needs, please input the following information: ")
    Sex_27 = str(input('Sex[F/M]: '))
    Height_27 = float(input('Height(m): '))
    Weight_27 = float(input('Weight(kg): '))
    Age_27 = float(input('Age: '))
    if Sex_27 == 'M' or Sex_27 == 'm':
        bmrmale_27 = float(88.36 + (13.4 * Weight_27) + (4.8 * (Height_27 * 100)) - (5.7 * Age_27))
    if Sex_27 == 'F' or Sex_27 == 'f':
        bmrfemale_27 = float(447.6 + (9.2 * Weight_27) + (3.1 * (Height_27 * 100)) - (4.3 * Age_27))

    print("[1]Sedentary")
    print("[2]Lightly active")
    print("[3]Moderately active")
    print("[4]Very active")
    print("[5]Extra active")
    Activity_27 = input('Please choose your activity level: ')

    if Activity_27 == '1' and (Sex_27 == 'M' or Sex_27 == 'm'):
        Calneeds1_27 = float(bmrmale_27 * 1.2)
        print('Your daily calorie needs is: ')
        print(Calneeds1_27)
    if Activity_27 == '1' and (Sex_27 == 'F' or Sex_27 == 'f'):
        Calneeds2_27 = float(bmrfemale_27 * 1.2)
        print('Your daily calorie needs is: ')
        print(Calneeds2_27)
    if Activity_27 == '2' and (Sex_27 == 'M' or Sex_27 == 'm'):
        Calneeds3_27 = float(bmrmale_27 * 1.375)
        print('Your daily calorie needs is: ')
        print(Calneeds3_27)
    if Activity_27 == '2' and (Sex_27 == 'F' or Sex_27 == 'f'):
        Calneeds4_27 = float(bmrfemale_27 * 1.375)
        print('Your daily calorie needs is: ')
        print(Calneeds4_27)
    if Activity_27 == '3' and (Sex_27 == 'M' or Sex_27 == 'm'):
        Calneeds5_27 = float(bmrmale_27 * 1.55)
        print('Your daily calorie needs is: ')
        print(Calneeds5_27)
    if Activity_27 == '3' and (Sex_27 == 'F' or Sex_27 == 'f'):
        Calneeds6_27 = float(bmrfemale_27 * 1.55)
        print('Your daily calorie needs is: ')
        print(Calneeds6_27)
    if Activity_27 == '4' and (Sex_27 == 'M' or Sex_27 == 'm'):
        Calneeds7_27 = float(bmrmale_27 * 1.725)
        print('Your daily calorie needs is: ')
        print(Calneeds7_27)
    if Activity_27 == '4' and (Sex_27 == 'F' or Sex_27 == 'f'):
        Calneeds8_27 = float(bmrfemale_27 * 1.725)
        print('Your daily calorie needs is: ')
        print(Calneeds8_27)
    if Activity_27 == '5' and (Sex_27 == 'M' or Sex_27 == 'm'):
        Calneeds9_27 = float(bmrmale_27 * 1.9)
        print('Your daily calorie needs is: ')
        print(Calneeds9_27)
    if Activity_27 == '5' and (Sex_27 == 'F' or Sex_27 == 'f'):
        Calneeds10_27 = float(bmrfemale_27 * 1.9)
        print('Your daily calorie needs is: ')
        print(Calneeds10_27)
    if Activity_27 > '5':
        print('Input invalid.')
    print("Disclamer: This is just a rough estimate and your actual calorie needs may vary depending on a variety of factors. It's always best to consult with a healthcare professional or registered dietitian to determine the best calorie intake for your individual needs and goals.")


```
