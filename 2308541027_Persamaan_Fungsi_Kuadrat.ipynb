```python
import cmath
import numpy as np
import matplotlib.pyplot as plt

print("QUADRATIC EQUATION GRAPHIC")
print()

# Set while loop
repetition_27 = True
while repetition_27 == True:

    # Input parameters for the quadratic equation
    print("Please input the parameters for the quadratic equation")
    a_27 = float(input('Input the value of a '))
    b_27 = float(input('Input the value of b '))
    c_27 = float(input('Input the value of c '))
    print()

    # Determine whether it is a quadratic function or not
    if a_27 == 0:
        print("It is not a quadratic formula.")
    
    else:
    
        print("Here's a few info about the parabola:")
        if a_27 > 0:
            print("The function curves upward.")
        if a_27 < 0:
            print("The function curves downward.")
    
        D_27 = b_27 * b_27 - (4 * a_27 * c_27)  # Define Discriminant 
    
        # Use quadratic formula to find the roots
        x1_27 = (-b_27 + cmath.sqrt(D_27)) / (2 * a_27)  # first root
        x2_27 = (-b_27 - cmath.sqrt(D_27)) / (2 * a_27)  # second root
    
        # Vertex parameters
        xp_27 = -b_27 / (2 * a_27) 
        yp_27 = a_27 * (xp_27 ** 2) + b_27 * xp_27 + c_27 
    
        if b_27 == 0:
            print("The vertex is at the center of the y-axis.")
        elif b_27 > 0 and a_27 > 0:
            print("The vertex is to the left of the y-axis.")
        elif b_27 < 0 and a_27 > 0:
            print("The vertex is to the right of the y-axis.")  # Displaying where the vertex is to the y-axis
        elif b_27 > 0 and a_27 < 0:
            print("The vertex is to the right of the y-axis.")
        elif b_27 < 0 and a_27 < 0:
            print("The vertex is to the left of the y-axis.")
    
        if c_27 > 0:
            print("The function intercepts the y-axis in the positive y region.")
        elif c_27 == 0:
            print("The function intercepts the y-axis at the center.")  # displaying the position of the intersection with the y-axis
        else:
            print("The function intercepts the y-axis in the negative y region.")
        
        # Displaying various info about the quadratic equation
        print("Vertex coordinate: ({:.2f},{:.2f})".format(xp_27, yp_27))
        print('Axis of symmetry: x = {:.2f}'.format(xp_27))
        print("Y-intercept: (0,{:.2f})".format(c_27))
    
        # Input the roots to quadratic form to plot (setting the base)
        x_27 = np.linspace(xp_27 - 4, xp_27 + 4, 100)
        y_27 = a_27 * (x_27 ** 2) + b_27 * x_27 + c_27
    
        # Parabola plot
        plt.plot(x_27, y_27, color='green', label='{:.2f}xÂ² + {:.2f}x + {:.2f}'.format(a_27, b_27, c_27))
        
        plt.axvline(x=xp_27, color='blue', linestyle='dashed', label='x = {:.2f}'.format(xp_27))  # axis of symmetry plot
        plt.legend(loc='upper left')  # Labels position
        plt.title("Quadratic Equation Graphic")
        plt.grid(True)  # grid plot
        plt.axvline(x=0, color='black')  # coordinate axes plot
        plt.axhline(y=0, color='black')
    
        # Extra elements based on discriminants (embellishments, I suppose)
        if D_27 > 0:
            print('The quadratic equation has two distinct real roots, namely x = {:.2f} and x = {:.2f}.'.format(x1_27.real, x2_27.real))
            print('Thus the parabola crosses the x-axis at two points.')
            plt.scatter([x1_27.real, x2_27.real], [0, 0], color='navy', s=60)
            plt.text(x1_27.real, 0, f'({x1_27.real:.2f}, 0)', ha='left', va='top', color='black')
            plt.text(x2_27.real, 0, f'({x2_27.real:.2f}, 0)', ha='right', va='top', color='black')

            plt.scatter([xp_27], [yp_27], color='red', s=60)
            plt.text(xp_27, yp_27, f'({xp_27:.2f}, {yp_27:.2f})', va='top', color='black')
    
            plt.show()
        
            Answer_27 = input('Do you want to try another problem?[Y/N]')
            if Answer_27.lower() == 'n':
                repetition_27 = False
        
        elif D_27 == 0:
            print('The quadratic equation has exactly one real root, namely x = {:.2f}.'.format(x1_27.real))
            print('Thus the parabola touches the x-axis at exactly one point.')
            plt.scatter([x1_27.real], [0], color='navy', s=60)
            plt.text(x1_27.real, 0, f'({x1_27.real:.2f}, 0)', ha='left', va='top', color='black')

            plt.scatter([xp_27], [yp_27], color='red', s=60)
            plt.text(xp_27, yp_27, f'({xp_27:.2f}, {yp_27:.2f})', va='top', color='black')
    
            plt.show()
            
            Answer_27 = input('Do you want to try another problem?[Y/N]')
            if Answer_27.lower() == 'n':
                repetition_27 = False
        
        elif D_27 < 0:
            print('The quadratic equation has two complex (non-real) roots, namely x = {:.2f} and x = {:.2f}.'.format(x1_27.real, x2_27.real))
            print('Thus the parabola does not intersect the x-axis at all.')
            plt.scatter([xp_27], [yp_27], color='red', s=60)
            plt.text(xp_27, yp_27, f'({xp_27:.2f}, {yp_27:.2f})', va='top', color='black')
    
            plt.show()
            
            Answer_27 = input('Do you want to try another problem?[Y/N]')
            if Answer_27.lower() == 'n':
                repetition_27 = False
```
